#Power BI Basics


Question 1: What is Power BI and why is it used in businesses?

- Answer:

  Power BI is a business intelligence and data visualization tool developed by Microsoft. It is used to analyze data, create interactive dashboards and reports, and support better business decision-making.

---

Question 2: Name and explain the three main components of Power BI.

- Answer:

1. **Power BI Desktop** – Used to connect, transform, model, and visualize data.
2. **Power BI Service** – Cloud platform used to publish, share, and collaborate on reports and dashboards.
3. **Power BI Mobile** – Mobile application used to view reports and dashboards on smartphones and tablets.

---

Question 3: Explain the Power BI workflow.

- Answer:

1. Connect to data sources.
2. Clean and transform data using Power Query.
3. Create relationships and data models.
4. Build reports and dashboards.
5. Publish reports to Power BI Service.
6. Share and refresh reports.

---

Question 4: List any four data cleaning tasks that can be performed in Power Query.

- Answer:

- Remove duplicate rows.
- Handle missing values.
- Change data types.
- Split or merge columns.

---

Question 5: Write step-by-step instructions to load the Real_Estate dataset.

- Answer:

1. Open **Power BI Desktop**.
2. Click **Home → Get Data**.
3. Select the dataset file (Excel/CSV).
4. Click **Open**.
5. Preview the data.
6. Click **Load** or **Transform Data** if cleaning is required.
7. The dataset is loaded into Power BI.

---

Question 6: Define Data View, Report View, and Model View. Explain the purpose of each view.

- Answer:

- **Data View:** Used to view and inspect table data.
- **Report View:** Used to create charts, reports, and dashboards.
- **Model View:** Used to create and manage relationships between tables.

Question 8: Split Owner Name to create two new columns as First Name and Last Name.

- Answer:

Steps:
1. Open **Power Query Editor**.
2. Select the **Owner Name** column.
3. Go to **Transform → Split Column → By Delimiter**.
4. Choose **Space** as the delimiter.
5. Click **OK**.
6. Rename the new columns as **First Name** and **Last Name**.
