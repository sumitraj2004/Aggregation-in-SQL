Question 1: What is Data Loading in Power BI and why is it considered the first step of analysis?

- Answer:

  Data Loading is the process of importing data from different sources into Power BI. It is considered the first step of analysis because reports and dashboards cannot be created until the data is loaded.

---

Question 2: Explain the difference between "Load" and "Transform Data" in Power BI.

- Answer:

**Load:** Imports data directly into Power BI without making changes.

**Transform Data:** Opens Power Query Editor to clean, filter, modify, and prepare the data before loading.

Question 3: What is a Fact Table and a Dimension Table? Give examples from the dataset.

- Answer:

**Fact Table:** Contains measurable values.

Example: **Confirmed_Cases, Recovered_Cases, Deaths, Active_Cases, Tests_Conducted, Hospitalized**

**Dimension Table:** Contains descriptive information.

Example: **Country, State, Date, Vaccination_Status**

Question 5: Identify and remove duplicate records based on Date, Country, and State.

- Answer:

**Steps:**
1. Open **Power Query Editor**.
2. Select **Date, Country, and State** columns.
3. Go to **Home → Remove Rows → Remove Duplicates**.
4. Click **Close & Apply**.

**Result:** No duplicate records were found in the dataset.

Question 6: Identify and replace null values in Vaccination_Status.

- Answer:

**Steps:**
1. Select the **Vaccination_Status** column.
2. Go to **Transform → Replace Values**.
3. Replace **null** with **"Unknown"**.
4. Click **Close & Apply**.

**Result:** The dataset contains **170 null values**, which should be replaced with **"Unknown"**.

Question 7: Create a new column to calculate Recovery Rate.

- Answer:

**DAX Formula:**

```DAX
Recovery Rate =
DIVIDE(
    coronavirus_2020_dataset[Recovered_Cases],
    coronavirus_2020_dataset[Confirmed_Cases],
    0
)
```

Question 8: Create a summarized table showing total confirmed cases by Country.

- Answer:

**DAX Formula:**

```DAX
Country Summary =
SUMMARIZE(
    coronavirus_2020_dataset,
    coronavirus_2020_dataset[Country],
    "Total Confirmed Cases",
    SUM(coronavirus_2020_dataset[Confirmed_Cases])
)
```
