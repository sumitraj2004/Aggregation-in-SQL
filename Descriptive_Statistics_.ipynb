2. Create a PivotTable to show Total Sales by Region.

   - Steps:

     1. Select the dataset.
     2. Go to Insert → PivotTable.
     3. Choose New Worksheet.
     4. Drag Region to Rows.
     5. Drag Sales to Values (Sum of Sales).

     This displays the Total Sales by Region.

3. Apply Conditional Formatting to highlight sales greater than 400.
   - Steps:

     1. Select the Sales column.
     2. Go to Home → Conditional Formatting → Highlight Cells Rules → Greater Than.
     3. Enter 400.
C    4. Choose a formatting color.
     Click OK.

5. Create a PivotChart showing Monthly Sales Trend.

   - Steps:

     1. Create a PivotTable.
     2. Drag Date to Rows and group by Months.
     3. Drag Sales to Values.
     4. Go to PivotTable Analyze → PivotChart.
     5. Select Line Chart.
