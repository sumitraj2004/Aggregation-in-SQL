#Logical Functions & Lookup

1. Write a formula to check if the marks in cell B2 are greater than or equal to 40. If yes, return "Pass", otherwise "Fail".
   - Formula:

      = IF(B2>=40,"Pass","Fail")

2. A company gives a bonus if Sales are above 40,000 and Attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not.
  - Formula:

     = IF(AND(B2>40000,C2>=90),"Bonus","No Bonus")

3. Write a nested IF formula.

   - Condition:

     Marks ≥ 90 → A
     Marks ≥ 75 → B
     Marks ≥ 50 → C
     Otherwise → Fail

     Formula:

     = IF(B2>=90,"A",IF(B2>=75,"B",IF(B2>=50,"C","Fail")))

4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

   - Formula (VLOOKUP):

     = VLOOKUP("E103",A2:C5,3,FALSE)

       Answer: Finance

       Explanation:
       VLOOKUP is used because the Employee ID is in the first column, and it returns the Department from the third column using an exact match.

5. A company has set the following commission structure. If a salesperson achieves sales of ₹45,000, use the VLOOKUP function to determine the commission rate applicable.

Formula:

=VLOOKUP(45000,A2:B5,2,TRUE)

Answer: 10%

7. Find the salary of EmpID = E105 using INDEX-MATCH Function.

   - Formula:

     = INDEX(B2:B5,MATCH("E105",A2:A5,0))

      Answer: 72000

8. Use the XLOOKUP function to find the Department of the employee with Emp ID = E303.

  - Formula:

     = XLOOKUP("E303",A2:A5,B2:B5)

      Answer: IT
