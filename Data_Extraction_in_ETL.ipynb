#Data Extraction in ETL

ETL Assignment

Question 1: Describe different types of data sources used in ETL with suitable examples.

- Answer:

  Common data sources used in ETL are:

- Databases (MySQL, SQL Server, Oracle)
- Flat Files (CSV, Excel, Text files)
- APIs (REST, SOAP)
- Cloud Storage (AWS S3, Azure Blob)
- Data Warehouses (Snowflake, BigQuery)

  These sources provide data for extraction into the ETL pipeline.

---

Question 2: What is data extraction? Explain its role in the ETL pipeline.

- Answer:

  Data extraction is the process of collecting data from different sources. It is the first step of ETL and provides raw data for transformation and loading into a target system.

---

Question 3: Explain the difference between CSV and Excel in terms of extraction and ETL usage.

- Answer:

| CSV | Excel |
|------|-------|
| Plain text file | Spreadsheet file |
| Faster to process | Slower due to formatting |
| Easy to automate | Supports formulas and multiple sheets |
| Commonly used in ETL | Used for reporting and analysis |

---

Question 4: Explain the steps involved in extracting data from a relational database.

- Answer:

1. Connect to the database.
2. Select the required tables.
3. Write SQL queries.
4. Extract the required data.
5. Validate the extracted data.
6. Store or load the data for further processing.

---

Question 5: Explain three common challenges faced during data extraction.

- Answer:

1. Poor data quality.
2. Large data volume.
3. Connection or network issues.

---

Question 6: What are APIs? Explain how APIs help in real-time data extraction.

- Answer:

  APIs (Application Programming Interfaces) allow applications to exchange data. They provide real-time data extraction by sending and receiving data instantly between systems.

---

Question 7: Why are databases preferred for enterprise-level data extraction?

- Answer:

  Databases provide high performance, better security, data consistency, scalability, and support for large volumes of structured data, making them suitable for enterprise-level ETL.

---

Question 8: What steps should an ETL developer take when extracting data from large CSV files (1GB+)?

- Answer:

- Read data in chunks.
- Filter only required columns.
- Validate data while extracting.
- Handle missing or duplicate records.
- Optimize memory usage.
- Load data in batches for better performance.
