Question 1: Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

- Answer:

Use the formula:

=BINOM.DIST(5,30,1/6,FALSE)

Result ≈ 0.1724 (17.24%)

Question 3: A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

- Answer:

The interval 45–55 is one standard deviation from the mean (50 ± 5).

According to the Empirical Rule, approximately 68% of the values lie within one standard deviation of the mean.

- Answer: Approximately 68%.

Question 4: What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

- Answer:

Standardization (Z-score) converts data into a common scale with a mean of 0 and a standard deviation of 1.

Formula:

Z = (X − μ) / σ

Where:
- X = Data value
- μ = Mean
- σ = Standard deviation

It is important because it allows comparison of values from different datasets, detects outliers, and improves the performance of many machine learning algorithms.

Question 5: What is Kurtosis and their type?

- Answer:

Kurtosis measures the peakedness and tail heaviness of a distribution.

Types:
- Mesokurtic – Normal distribution.
- Leptokurtic – High peak with heavy tails.
- Platykurtic – Flat peak with light tails.

Question 6: Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

- Answer:

A fair die has six outcomes, and each outcome has an equal probability (1/6). Therefore, the outcomes follow a discrete uniform distribution.

Question 7: Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5.

- Answer:

Excel Formula:

=1-BINOM.DIST(7,15,0.5,TRUE)

Result ≈ 0.3036 (30.36%)

Question 8: How does log transformation help in stabilizing variance and making data more normally distributed?

- Answer:

Log transformation reduces the effect of large values, stabilizes variance, reduces right skewness, and makes data closer to a normal distribution. It also improves the accuracy of statistical analysis and machine learning models.
