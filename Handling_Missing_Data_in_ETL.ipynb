#Data Quality and Validation in ETL



Question 1: Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

- Answer:

  Data Quality refers to the accuracy, completeness, consistency, validity, and reliability of data in an ETL pipeline. It is more than data cleaning because it also includes validating, monitoring, and maintaining high-quality data throughout the ETL process.

Question 2: Explain why poor data quality leads to misleading dashboards and incorrect decisions.

- Answer:

  Poor data quality results in inaccurate reports, incorrect KPIs, duplicate records, missing values, and inconsistent information. This can lead to misleading dashboards and poor business decisions.

Question 3: What is duplicate data? Explain three causes in ETL pipelines.

- Answer:

 Duplicate data refers to repeated records representing the same information.

Common causes:
- Multiple data sources.
- Repeated ETL execution.
- Missing primary key or validation rules.

Question 4: Differentiate between exact, partial, and fuzzy duplicates.

- Answer:

| Type | Description |
|------|-------------|
| Exact Duplicate | All values are exactly the same. |
| Partial Duplicate | Only some fields are duplicated. |
| Fuzzy Duplicate | Similar records with spelling or formatting differences. |

Question 5: Why should data validation be performed during transformation rather than after loading?

- Answer:

  Data validation during transformation detects and corrects errors before loading, improving data quality and preventing incorrect data from entering the target database.

Question 7: Write an SQL query on Sales_Transactions to list all duplicate keys and their counts using the business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount).

- Answer:

```sql
SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;
```

Question 8: Enforcing Referential Integrity

Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with Customers_Master and write a query to detect such violations.

- Answer:

**Violating Customer_IDs:** **C105** and **C106**

```sql
SELECT s.Customer_ID
FROM Sales_Transactions s
LEFT JOIN Customers_Master c
ON s.Customer_ID = c.CustomerID
WHERE c.CustomerID IS NULL;
```
