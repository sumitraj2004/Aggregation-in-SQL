#Handling Missing Data in ETL

Question 1: What are the most common reasons for missing data in ETL pipelines?

- Answer:

The most common reasons are:
- Data entry errors
- System or network failures
- Missing values from source systems
- Data corruption
- Incomplete data collection

---

Question 2: Why is blindly deleting rows with missing values considered a bad practice in ETL?

- Answer:

  Blindly deleting rows may remove important information, reduce the dataset size, introduce bias, and affect the accuracy of analysis and machine learning models.

---

Question 3: Explain the difference between Listwise deletion and Column deletion. Also mention one scenario where each is appropriate.

- Answer:

  **Listwise Deletion:**
    Removes entire rows containing missing values.

  **Use Case:** Suitable when only a few rows have   missing values.

**Column Deletion:**
 Removes an entire column with many missing values.

**Use Case:** Suitable when a column has a very high percentage of missing values and is not important.

---

Question 4: Why is median imputation preferred over mean imputation for skewed data such as income?

- Answer:

  Median is preferred because it is not affected by extreme values or outliers, making it more accurate for skewed data like income.

---

Question 5: What is forward fill and in what type of dataset is it most useful?

- Answer:

   Forward fill replaces missing values with the previous valid value. It is most useful for time-series data such as stock prices, sales, and sensor data.

---

Question 6: Why should flagging missing values be done before imputation in an ETL workflow?

- Answer:

  Flagging missing values preserves information about which values were originally missing and helps improve data analysis and machine learning models.

---

Question 7: Consider a scenario where income is missing for many customers. How can this missingness itself provide business insights?

- Answer:

  Missing income values may indicate incomplete customer records, poor data collection, or customers unwilling to share information. This insight can help improve business processes and data quality.
