#Publishing & Sharing Reports in Power BI

Publishing & Sharing Reports in Power BI

Question 1: Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop.

Answer:

Publishing and sharing reports allow users to access, collaborate, and make business decisions using the latest data. Dashboards built only in Power BI Desktop cannot be shared or accessed online.

---

Question 2: Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.

Answer:

**Power BI Desktop** is used to create reports, transform data, and build data models.

**Power BI Service** is the cloud platform used to publish, share, refresh, and collaborate on reports and dashboards.

---

Question 3: What does publishing a report mean in Power BI? List any two outcomes that occur after a report is published.

Answer:

Publishing a report means uploading it from Power BI Desktop to Power BI Service.

Outcomes:
- Reports become available online.
- Reports can be shared with other users.

---

Question 4: List and explain the steps involved in publishing a report from Power BI Desktop to Power BI Service in the correct sequence.

Answer:

1. Create the report in Power BI Desktop.
2. Save the report.
3. Sign in to Power BI Desktop.
4. Click **Publish**.
5. Select the target workspace.
6. Click **Select**.
7. Open the report in Power BI Service.

---

Question 5: Differentiate between Report and Dashboard in Power BI based on structure and usage.

Answer:

| Report | Dashboard |
|--------|-----------|
| Multi-page | Single-page |
| Built in Power BI Desktop | Built in Power BI Service |
| Used for detailed analysis | Used for summary and monitoring |
| Contains multiple visuals | Contains pinned visuals |

---

Question 6: Explain the difference between the following three sharing methods in Power BI Service: Direct Report Sharing, Workspace Access, and Power BI Apps. Also mention one use case for each.

Answer:

**Direct Report Sharing**
- Shares a single report with specific users.
- **Use Case:** Share a report with one manager.

**Workspace Access**
- Allows team members to collaborate on reports.
- **Use Case:** Analytics team working on reports together.

**Power BI Apps**
- Distributes reports to a large audience in a controlled way.
- **Use Case:** Company-wide sales dashboard.

---

Question 7: Describe the four workspace roles available in Power BI and explain how access control is managed using these roles.

Answer:

- **Admin:** Full control over the workspace.
- **Member:** Create, edit, and publish content.
- **Contributor:** Add and update reports.
- **Viewer:** View reports only.

Access control is managed by assigning users the appropriate workspace role.

---

Question 8: Why are Power BI Apps considered more secure and scalable than direct sharing? Explain this from a governance and enterprise usage perspective.

Answer:

Power BI Apps provide centralized management, version control, secure distribution, and easier updates. They are ideal for enterprise-wide reporting because administrators can manage permissions from one place.

---

Question 9: Explain Row-Level Security (RLS) in Power BI. How does RLS ensure that different users see different data using the same report?

Answer:

Row-Level Security (RLS) restricts data access based on user roles. Users view the same report but only see the rows they are authorized to access.

Example:
A North Region manager sees only North Region sales, while a South Region manager sees only South Region sales.

---

Question 10: Assume an organization shares reports without using Apps, workspace roles, or RLS. Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain how Power BI best practices help prevent them.

Answer:

Risks:
- Unauthorized users may access confidential data.
- Incorrect report versions may be shared.
- Sensitive business information may be misused.

Best Practices:
- Use Power BI Apps for secure distribution.
- Assign Workspace Roles for controlled access.
- Implement Row-Level Security (RLS) to restrict data visibility.
