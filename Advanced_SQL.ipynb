Question 1: What is a Common Table Expression (CTE), and how does it improve SQL query readability?

- Answer:

  A Common Table Expression (CTE) is a temporary result set created using the `WITH` clause. It improves SQL query readability by breaking complex queries into smaller, reusable parts, making them easier to read, write, and maintain.



Question 2: Why are some views updatable while others are read-only? Explain with an example.

- Answer:

  A view is **updatable** if it is based on a single table without aggregate functions, GROUP BY, DISTINCT, or JOINs. A view becomes **read-only** when it contains these operations.

  Example:

```sql
CREATE VIEW EmployeeView AS
SELECT EmployeeID, Name
FROM Employees;
```

  This view is updatable because it is based on a single table.

Question 6: Write a CTE to calculate the total revenue for each product (Revenue = Price × Quantity), and return only products where revenue > 3000.

- Answer:

```sql
WITH RevenueCTE AS (
    SELECT
        p.ProductID,
        p.ProductName,
        p.Price * s.Quantity AS Revenue
    FROM Products p
    JOIN Sales s
    ON p.ProductID = s.ProductID
)
SELECT *
FROM RevenueCTE
WHERE Revenue > 3000;
```

Question 7: Create a view named vw_CategorySummary that shows Category, TotalProducts, and AveragePrice.

- Answer:

```sql
CREATE VIEW vw_CategorySummary AS
SELECT
    Category,
    COUNT(*) AS TotalProducts,
    AVG(Price) AS AveragePrice
FROM Products
GROUP BY Category;
```

Question 8: Create an updatable view containing ProductID, ProductName, and Price. Then update the price of ProductID = 1 using the view.

- Answer:

```sql
CREATE VIEW vw_ProductDetails AS
SELECT ProductID, ProductName, Price
FROM Products;
```

```sql
UPDATE vw_ProductDetails
SET Price = 1500
WHERE ProductID = 1;
```

Question 9: Create a stored procedure that accepts a category name and returns all products belonging to that category.

- Answer:

```sql
CREATE PROCEDURE GetProductsByCategory
    @CategoryName VARCHAR(50)
AS
BEGIN
    SELECT *
    FROM Products
    WHERE Category = @CategoryName;
END;
```

 Example:

```sql
EXEC GetProductsByCategory 'Electronics';
```

Question 10: Create an AFTER DELETE trigger on the Products table that archives deleted product rows into a new table ProductArchive. The archive should store ProductID, ProductName, Category, Price, and DeletedAt timestamp.

- Answer:

```sql
CREATE TABLE ProductArchive (
    ProductID INT,
    ProductName VARCHAR(100),
    Category VARCHAR(50),
    Price DECIMAL(10,2),
    DeletedAt DATETIME
);
```

```sql
CREATE TRIGGER trg_ProductArchive
ON Products
AFTER DELETE
AS
BEGIN
    INSERT INTO ProductArchive
    (ProductID, ProductName, Category, Price, DeletedAt)
    SELECT
        ProductID,
        ProductName,
        Category,
        Price,
        GETDATE()
    FROM deleted;
END;
```
