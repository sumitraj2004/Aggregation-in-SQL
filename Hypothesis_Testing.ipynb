#Hypothesis Testing

Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

- Answer:

The null hypothesis (H₀) is the default assumption that there is no significant difference or effect. It is important because it provides a basis for statistical testing and helps determine whether there is enough evidence to reject the claim.

---

Question 2: What does the significance level (α) represent in hypothesis testing?

- Answer:

The significance level (α) is the maximum probability of making a Type I error (rejecting a true null hypothesis). The most commonly used significance level is 0.05 (5%).

Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.

- Answer:

One-tailed Test:
- Tests for a difference in one direction only.
- Example: Testing whether the average salary is greater than ₹50,000.

Two-tailed Test:
- Tests for a difference in both directions.
- Example: Testing whether the average salary is different from ₹50,000.

Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

- Answer:

Given:
- μ = 10
- x̄ = 12
- s = 3
- n = 9
- α = 0.05

Test Statistic:

t = (12 − 10) / (3/√9)

= 2 / 1

= 2

Degrees of Freedom = 8

Critical t-value (α = 0.05, two-tailed) = ±2.306

Since 2 < 2.306, fail to reject H₀.

Conclusion:
There is insufficient evidence to reject the company's claim.

Question 6: When should you use a Z-test instead of a t-test?

- Answer:

A Z-test is used when the population standard deviation is known or the sample size is large (n ≥ 30). A t-test is used when the population standard deviation is unknown and the sample size is small.

Question 8: A company wants to test if product preference is independent of gender. At α = 0.05, test independence.

- Answer:

Use the Chi-Square Test of Independence.

H₀: Product preference is independent of gender.

H₁: Product preference is dependent on gender.

Since the observed frequencies differ noticeably from the expected frequencies, reject H₀.

Conclusion:
Product preference is not independent of gender. There is a significant association between gender and product preference at α = 0.05.
