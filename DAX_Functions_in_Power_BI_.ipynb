#Advanced Visuals in PowerBI

Power BI Assignment – Filters, Slicers, Drill-Down & Drill-Through

Question 1: What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?

Answer:

- **Basic Filter:** Filters data by selecting one or more values.
- **Advanced Filter:** Filters data using conditions (e.g., Greater Than, Contains, Between).
- **Slicer:** A visual filter that allows users to interactively filter report data.

Question 4: What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?

Answer:

- **Visual-Level Filter:** Applies only to one visual.
- **Page-Level Filter:** Applies to all visuals on the current page.
- **Report-Level Filter:** Applies to all pages in the report.

Question 10: Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.

Answer:

- **Drill-Down:** Explores different levels within the same chart.
- **Drill-Through:** Opens another report page with detailed information.

**Example:** Drill down from **Country → Region → Category**, then drill through to view detailed sales for a selected product.

Question 13: What is the "Keep All Filters" option in Drill-Through? How does it affect user interaction?

Answer:

**Keep All Filters** passes all existing filters to the Drill-Through page, allowing users to see detailed data with the same filter context.
