#DAX Functions in Power BI

Amazon Store Sales Dataset – Power BI Assignment

Question 1: Create a calculated column **TotalPrice** using **Quantity** and **Unit Price**.

- Answer:

```DAX
TotalPrice =
'Amazon Store Sales'[Quantity] *
'Amazon Store Sales'[UnitPrice]
```

---

Question 2: Create a calculated column **DiscountedPrice**.

- Answer:

```DAX
DiscountedPrice =
'Amazon Store Sales'[TotalPrice] *
(1 - 'Amazon Store Sales'[Discount] / 100)
```

---

Question 3: Create a measure **Average Order Value (AOV)**.

- Answer:

```DAX
Average Order Value =
DIVIDE(
    SUM('Amazon Store Sales'[SalesAmount]),
    DISTINCTCOUNT('Amazon Store Sales'[OrderID])
)
```

---

Question 4: Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order Date.

- Answer:

```DAX
YTD Sales =
TOTALYTD(
    SUM('Amazon Store Sales'[SalesAmount]),
    'Amazon Store Sales'[OrderDate]
)
```

---

Question 5: Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling product ranked first.

- Answer:

```DAX
Product Rank =
RANKX(
    ALL('Amazon Store Sales'[ProductID]),
    CALCULATE(SUM('Amazon Store Sales'[SalesAmount])),
    ,
    DESC,
    DENSE
)
```

---

Question 6: Create a calculated column classifying orders as **"High Value"** if Sales > 2000, else **"Low Value"**.

- Answer:

```DAX
Order Category =
IF(
    'Amazon Store Sales'[SalesAmount] > 2000,
    "High Value",
    "Low Value"
)
```

---

Question 7: Create a measure to calculate **total sales from the Online channel only**.

- Answer:

```DAX
Online Sales =
CALCULATE(
    SUM('Amazon Store Sales'[SalesAmount]),
    'Amazon Store Sales'[SalesChannel] = "Online"
)
```

---

Question 8: Create a measure called **Date Difference** that calculates the number of days between the latest Order Date and today's date using the DATEDIFF function.

- Answer:

```DAX
Date Difference =
DATEDIFF(
    MAX('Amazon Store Sales'[OrderDate]),
    TODAY(),
    DAY
)
```
