#Probability Basics


Question 1: A die is rolled. What is the probability of getting:

           - (a) An even number

               Even numbers on a die are **2, 4, and 6**.

               \[
               P(\text{Even})=\frac{3}{6}=\frac{1}{2}
               \]

               **Probability = 1/2**

          (b) A number greater than 4

            - Numbers greater than 4 are **5 and 6**.

              \[
              P(>4)=\frac{2}{6}=\frac{1}{3}
              \]

              **Probability = 1/3**

              



Question 2: In a class of 50 students:

- 20 like Mathematics (M)
- 15 like Science (S)
- 5 like both subjects

**What is the probability that a student chosen at random likes Mathematics or Science?**

   - Answer:

Using the formula:

\[
P(M\cup S)=\frac{M+S-(M\cap S)}{Total}
\]

\[
=\frac{20+15-5}{50}
=\frac{30}{50}
=\frac{3}{5}
=0.6
\]

**Probability = 0.6 (60%)**

Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?

- Answer:

After removing one red ball:

- Remaining red balls = 2
- Total remaining balls = 4

\[
P=\frac{2}{4}=\frac{1}{2}
\]

**Probability = 1/2**

Question 5: The average height of 1000 students is 160 cm. A sample of 100 students shows an average height of 158 cm. Find the sampling error.

 - Answer:

Sampling Error = Sample Mean − Population Mean

= 158 − 160

= **−2 cm**

**Sampling Error = −2 cm**

Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?

 - Answer:

Formula:

\[
SEM=\frac{\sigma}{\sqrt{n}}
\]

\[
=\frac{5000}{\sqrt{100}}
=\frac{5000}{10}
=500
\]

**Standard Error (SEM) = ₹500**

Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?

 - Answer:

\[
P=\frac{4}{52}\times\frac{3}{51}
=\frac{12}{2652}
=\frac{1}{221}
\]

**Probability = 1/221**

Question 9: A factory produces bulbs with a 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?

 - Answer:

Probability of one non-defective bulb = **0.98**

\[

=0.904
\]
