#Pivot Table & Charts

1. What is the purpose of a PivotTable in Excel?
   - A PivotTable is used to summarize, analyze,      organize, and present large amounts of data. It  helps quickly calculate totals, averages, counts, and compare data.

2. Create a PivotTable to show Total Sales by Region?
   - Steps:

     1. Select the dataset.
     2. Go to Insert → PivotTable.
     3. Choose New Worksheet.
     4. Drag Region to Rows.
     5. Drag Sales to Values (Sum of Sales).

     This displays the Total Sales by Region.

3. Apply Conditional Formatting to highlight sales greater than 400.
   -Steps:

     1. Select the Sales column.
     2. Go to Home → Conditional Formatting → Highlight Cells Rules → Greater Than.
     3. Enter 400.
     4. Choose a formatting color.
     5. Click OK.

4. Explain the difference between Sorting and Filtering with an example from this dataset?
   - Sorting arranges data in ascending or descending order.
     Filtering displays only the records that meet specific conditions.

     Example:

     Sort Sales from Largest to Smallest.
     Filter Region = West to show only West region records.

5. Create a PivotChart showing Monthly Sales Trend.
  - Steps:

     1. Create a PivotTable.
     2. Drag Date to Rows and group by Months.
     3. Drag Sales to Values.
     4. Go to PivotTable Analyze → PivotChart.
     5. Select Line Chart.

6. Use a Slicer to filter data by Product and observe changes in sales ?
  
   - Steps:

     1. Select the PivotTable.
     2. Go to PivotTable Analyze → Insert Slicer.
     3. Select Product.
     4. Click OK.
     5. Use the slicer to filter products and observe the sales.

7. Why are Dashboards important for decision-making in business analysis?
   - Dashboards present data visually using charts, PivotTables, and slicers. They help users monitor performance, identify trends, compare data, and make faster business decisions.

9. Group dates in the PivotTable into Months and calculate the average sales per month.
  - Steps:

     1. Create a PivotTable.
     2. Drag Date to Rows.
     3. Right-click any date → Group → Months.
     4. Drag Sales to Values.
     5. Change Sum of Sales to Average
